```python
lst = df.columns
lst = lst[2:]
```

```python

fig, ax = plt.subplots(3,4,figsize=(15,8))
k = 0
try:
    for i in range(3):
        for j in range(4):
            plt.subplots_adjust(wspace=0.3,hspace=0.3)
            sns.countplot(data=df,x=lst[k],ax=ax[i,j])
            k +=1
except IndexError:
    pass
```
