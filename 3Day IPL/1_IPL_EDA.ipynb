```python
delivery = pd.read_csv("deliveries.csv")
delivery.head(2)
```

```python
match = pd.read_csv("matches.csv")
match.head(2)
```

```python
print(match.shape,delivery.shape)
```

```python
ipl = pd.merge(match,delivery,left_on="id",right_on="match_id",how = "outer")
```

```python
ipl.shape
```

```python
ipl.columns
```

```python
# we need [session, batting_team, batsman, batsman_run]

ipl_y = ipl.loc[:,['season', 'batting_team', 'batsman', 'batsman_runs']]
```

```python
ipl_y.info()
```

```python
ipl_y.sample(10)
```

```python
yellow = ipl_y.groupby(["season","batsman"])['batsman_runs'].sum().sort_values(ascending = False).reset_index()
```

```python
final = yellow.drop_duplicates('season',keep="first")
```

```python
final = final.sort_values('season').reset_index()
```
