```python
# find the most destrictive death over batsman in the history of the IPL
# need to find the strik rate of the betsman 
# formula  = (total run/total ball palyed)*100

# i have to also filter the true batsman so i pick above 200 ball played player 
```

```python
delivery.sample(2)
```

```python
# strick rate 



sr = delivery.groupby('batsman')
sr["batsman_runs"].sum().sort_values(ascending = False)
```

```python
sr["ball"].sum().sort_values(ascending = False)
```

```python
run = sr[["batsman_runs","ball"]].sum()
```

```python
run["strick"] = (run['batsman_runs']/run['ball'])*100
```

```python
run.sort_values(by = "strick", ascending=False).head(10)
```

```python
run[run['ball']>200].sort_values(by = "strick",ascending=False).head(10)
```
